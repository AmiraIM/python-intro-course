```python
car_nonans = car.dropna() #remove cars with car_nonans
car_nonans_num = car_nonans.select_dtypes(include='number')

car_nonans_num.head()
```

```python
carplot=sns.pairplot(car_nonans_num)
```

```python
#Apply k-means with a proper selection of k

from sklearn.cluster import KMeans
group = KMeans(n_clusters=2).fit_predict(car_nonans_num)
car_nonans_num['kMeans'] = group
car_k = sns.pairplot(car_nonans_num, hue="kMeans")
```
